## Genetic Algorithms on Geographic Datasets: Italian Cities Case Study

**Geographic Dataset**: Real coordinates of Italian cities

**Additional Challenges**: Non-uniform spatial distribution, complex geometry


- **Path Length Convergence**: Best + average across generations
- **Diversity Monitoring**: Maintaining genetic variability


```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({
    'figure.figsize': (12, 10),
    'font.size': 12,
    'font.family': 'serif',
    'axes.grid': True,
    'grid.alpha': 0.3
})

print("TSP ANALYSIS - ITALIAN CITIES DATASET")
print("=" * 50)

# Load Italian cities dataset
try:
    cities = np.loadtxt('cities_ita.dat')
    x_cities = cities[:, 0]
    y_cities = cities[:, 1]
    print(f"Italian cities dataset loaded: {len(cities)} cities")
except FileNotFoundError:
    print("File cities_ita.dat not found")
    cities = x_cities = y_cities = np.array([])

# Load optimal path
try:
    path = np.loadtxt('global_best_path_ita.dat')
    x_path = path[:, 0]
    y_path = path[:, 1]
    print(f"Optimal path loaded: {len(path)} points")
except FileNotFoundError:
    print("File global_best_path_ita.dat not found")
    path = x_path = y_path = np.array([])

if len(cities) > 0:
    # Morphological analysis of Italian territory
    print(f"\nITALIAN TERRITORY MORPHOLOGICAL ANALYSIS:")
    
    x_span = np.max(x_cities) - np.min(x_cities)
    y_span = np.max(y_cities) - np.min(y_cities)
    aspect_ratio = y_span / x_span
    area_convex = x_span * y_span
    
    print(f"   Number of cities: {len(cities)}")
    print(f"   Aspect ratio (L/W): {aspect_ratio:.3f}")
    print(f"   Bounding box area: {area_convex:.4f}")
    
    # Spatial distribution analysis
    center_x, center_y = np.mean(x_cities), np.mean(y_cities)
    distances_from_center = np.sqrt((x_cities - center_x)**2 + (y_cities - center_y)**2)
    mean_dist_center = np.mean(distances_from_center)
    std_dist_center = np.std(distances_from_center)
    
    print(f"\nSPATIAL DISTRIBUTION:")
    print(f"   Geometric center: ({center_x:.4f}, {center_y:.4f})")
    print(f"   Mean distance from center: {mean_dist_center:.4f} ± {std_dist_center:.4f}")
    print(f"   Dispersion coefficient: {std_dist_center/mean_dist_center:.3f}")
    
    # Distance matrix calculation
    print(f"\nDISTANCE MATRIX CALCULATION...")
    distance_matrix = np.zeros((len(cities), len(cities)))
    
    for i in range(len(cities)):
        for j in range(i+1, len(cities)):
            dist = np.sqrt((x_cities[i] - x_cities[j])**2 + (y_cities[i] - y_cities[j])**2)
            distance_matrix[i,j] = distance_matrix[j,i] = dist
    
    # Distance statistics
    distances = distance_matrix[np.triu_indices(len(cities), k=1)]
    mean_distance = np.mean(distances)
    std_distance = np.std(distances)
    min_distance = np.min(distances[distances > 0])
    max_distance = np.max(distances)
    


if len(path) > 0:
    # Optimal path analysis
    print(f"\nOPTIMAL PATH ANALYSIS:")
    
    # Calculate path statistics
    segment_lengths = []
    for i in range(len(path)):
        next_i = (i + 1) % len(path)
        dx = x_path[next_i] - x_path[i]
        dy = y_path[next_i] - y_path[i]
        segment_length = np.sqrt(dx**2 + dy**2)
        segment_lengths.append(segment_length)
    
    total_length = sum(segment_lengths)
    mean_segment = np.mean(segment_lengths)
    std_segment = np.std(segment_lengths)
    
    print(f"   Total path length: {total_length:.4f}")
    print(f"   Number of segments: {len(segment_lengths)}")
    print(f"   Mean segment length: {mean_segment:.4f} ± {std_segment:.4f}")
    print(f"   Segment range: [{min(segment_lengths):.4f}, {max(segment_lengths):.4f}]")
    
    # Efficiency vs theoretical bounds
    if 'mst_weight' in locals() and mst_weight is not None:
        efficiency_mst = mst_weight / total_length
        efficiency_christofides = christofides_bound / total_length
        
        print(f"   Efficiency vs MST: {efficiency_mst:.3f} ({efficiency_mst*100:.1f}%)")
        print(f"   Efficiency vs Christofides: {efficiency_christofides:.3f} ({efficiency_christofides*100:.1f}%)")
        print(f"   Gap from theoretical optimum: {((total_length - mst_weight)/mst_weight)*100:.2f}%")
    
    # Crossing detection analysis
    def line_segments_intersect(p1, p2, p3, p4):
        def ccw(A, B, C):
            return (C[1]-A[1]) * (B[0]-A[0]) > (B[1]-A[1]) * (C[0]-A[0])
        return ccw(p1,p3,p4) != ccw(p2,p3,p4) and ccw(p1,p2,p3) != ccw(p1,p2,p4)
    
    crossings = 0
    path_points = list(zip(x_path, y_path))
    n = len(path_points)
    
    for i in range(n):
        for j in range(i+2, n):
            if j == n-1 and i == 0:
                continue
            if line_segments_intersect(path_points[i], path_points[(i+1) % n],
                                     path_points[j], path_points[(j+1) % n]):
                crossings += 1
    
    print(f"   Crossings detected: {crossings}")
    print(f"   Topology: {'PLANAR (optimal)' if crossings == 0 else 'NON-PLANAR (suboptimal)'}")
    
    # Geographic coherence analysis
    geographic_jumps = 0
    for i in range(len(segment_lengths)):
        if segment_lengths[i] > mean_segment + 2*std_segment:
            geographic_jumps += 1
    
    coherence_score = 1 - (geographic_jumps / len(segment_lengths))
    print(f"   Excessive geographic jumps: {geographic_jumps}")
    print(f"   Geographic coherence score: {coherence_score:.3f}")

# Main visualization: Italian TSP solution
if len(cities) > 0 and len(path) > 0:
    fig, ax = plt.subplots(1, 1, figsize=(12, 10))
    
    # Plot cities
    scatter = ax.scatter(x_cities, y_cities, c=range(len(cities)), cmap='viridis', 
                         s=100, alpha=0.8, edgecolors='black', linewidth=1, 
                         label=f'Cities ({len(cities)})', zorder=3)
    
    # Plot optimal path
    ax.plot(x_path, y_path, 'r-', linewidth=3, alpha=0.8, label='Optimal Path', zorder=2)
    ax.plot(x_path, y_path, 'ro', markersize=6, alpha=0.7, zorder=2)
    ax.plot([x_path[-1], x_path[0]], [y_path[-1], y_path[0]], 'r-', linewidth=3, alpha=0.8)
    
    # Highlight start/end point
    ax.plot(x_path[0], y_path[0], 'go', markersize=12, markeredgecolor='black', 
            markeredgewidth=2, label='Start/End', zorder=4)
    
    # Add colorbar
    cbar = plt.colorbar(scatter, ax=ax)
    cbar.set_label('City Index', rotation=270, labelpad=15)
    
    ax.set_xlabel('X Coordinate')
    ax.set_ylabel('Y Coordinate')
    ax.set_title(f'TSP Italy - {len(cities)} Cities\nPath Length: {total_length:.4f}', 
                 fontweight='bold', fontsize=14)
    ax.legend()
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal')

plt.tight_layout()
plt.savefig('tsp_italy_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Load and analyze path length evolution data
try:
    evolution_data = np.loadtxt('fitness_history_ita.dat')
    generations = evolution_data[:, 0]
    path_lengths = evolution_data[:, 1]  # This is actually L2 path length, not fitness
    
    data_loaded = True
except FileNotFoundError:
    print("File fitness_history_ita.dat not found")
    print("Run the TSP GA algorithm first to generate evolution data")
    data_loaded = False

if data_loaded:
    # Create path length evolution visualization
    fig, ax1 = plt.subplots(1, 1, figsize=(12, 8))
    
    # Main path length evolution plot
    ax1.plot(generations, path_lengths, 'b-', linewidth=2, alpha=0.8, label='Best Path Length (L2)')
    
    # Add moving average for trend analysis
    if len(path_lengths) > 100:
        window_size = max(10, len(path_lengths) // 50)
        moving_avg = np.convolve(path_lengths, np.ones(window_size)/window_size, mode='valid')
        moving_gen = generations[window_size-1:]
        ax1.plot(moving_gen, moving_avg, 'r--', linewidth=2, alpha=0.7,
                label=f'Moving Average (window={window_size})')
    
    # Find and highlight the best (minimum) path length
    best_idx = np.argmin(path_lengths)
    best_generation = generations[best_idx]
    best_path_length = path_lengths[best_idx]
    
    # Add arrow pointing to the best solution
    ax1.annotate(f'Best Solution\nGen: {int(best_generation)}\nLength: {best_path_length:.4f}',
                xy=(best_generation, best_path_length),
                xytext=(best_generation + len(generations)*0.2, best_path_length + (np.max(path_lengths) - np.min(path_lengths))*0.3),
                arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.2', 
                              color='red', lw=2),
                fontsize=12, fontweight='bold',
                bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7),
                ha='center')
    
    # Highlight the best point
    ax1.plot(best_generation, best_path_length, 'ro', markersize=10, 
             markeredgecolor='black', markeredgewidth=2, zorder=5)
    
    ax1.set_xlabel('Generation')
    ax1.set_ylabel('Path Length (L2 Distance)')
    ax1.set_title('TSP Genetic Algorithm - Path Length Evolution', fontweight='bold', fontsize=14)
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('path_length_evolution_italy.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Convergence analysis
    if len(path_lengths) > 50:
        print(f"\nCONVERGENCE ANALYSIS:")
        
        # Find convergence point
        improvement_rate = np.diff(path_lengths) * -1
        convergence_threshold = np.std(improvement_rate) * 0.1
        
        convergence_gen = None
        if len(improvement_rate) > 100:
            for i in range(len(improvement_rate)-50, 50, -1):
                if np.mean(np.abs(improvement_rate[i:i+50])) > convergence_threshold:
                    convergence_gen = i
                    break
        
        if convergence_gen:
            print(f"   Convergence around generation: {convergence_gen}")
            print(f"   Path length at convergence: {path_lengths[convergence_gen]:.6f}")
            print(f"   Further improvement after convergence: {(path_lengths[convergence_gen] - path_lengths[-1]):.6f}")
                
        # Algorithm efficiency metrics
        total_improvement = path_lengths[0] - path_lengths[-1]
        efficient_gens = np.sum(improvement_rate > np.mean(improvement_rate))
        
        # Calculate optimization efficiency
        print(f"\nOPTIMIZATION EFFICIENCY:")
        print(f"   Initial L2 distance: {path_lengths[0]:.6f}")
        print(f"   Final L2 distance: {path_lengths[-1]:.6f}")
        print(f"   Total reduction: {total_improvement:.6f}")
        print(f"   Best solution found at generation: {int(best_generation)}")

print(f"\nPath length evolution analysis completed")
```

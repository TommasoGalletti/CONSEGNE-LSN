# Lezione 10.1 - TSP Optimization

## Implementation of Genetic Algorithms for Complex Geometries

### Objective

This lesson presents the evolution of genetic algorithms developed in Lesson 9

## Dataset and Experimental Configuration

### Problem Characteristics

The dataset represents a **standard geometric configuration** for TSP benchmarks:

- **Distribution**: Cities uniformly distributed in square domain [0,1]×[0,1]
- **Dimensionality**: Two-dimensional problem
- **Complexity**: Solution space = (N-1)!/2 possible paths
- **Objective**: Minimization of total distance L² = Σ||(x_i - x_{i+1})||²

### Performance Metrics

- **Solution Quality**: Path Distance
- **Convergence Speed**: Generations needed for convergence

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({
    'figure.figsize': (12, 10),
    'font.size': 12,
    'font.family': 'serif',
    'axes.grid': True,
    'grid.alpha': 0.3
})

print("TSP ANALYSIS - SQUARE CONFIGURATION")
print("=" * 50)

# Load city positions
try:
    cities = np.loadtxt('cities.dat')
    x_cities = cities[:, 0]
    y_cities = cities[:, 1]
    print(f"Dataset loaded: {len(cities)} cities")
except FileNotFoundError:
    print("File cities.dat not found")
    cities = x_cities = y_cities = np.array([])

# Load optimal path
try:
    path = np.loadtxt('global_best_path.dat')
    x_path = path[:, 0]
    y_path = path[:, 1]
    print(f"Optimal path loaded: {len(path)} points")
except FileNotFoundError:
    print("File global_best_path.dat not found")
    path = x_path = y_path = np.array([])

if len(cities) > 0:
    # Geometric analysis of dataset
    print(f"\nDATASET GEOMETRIC CHARACTERISTICS:")
    
    x_range = np.max(x_cities) - np.min(x_cities)
    y_range = np.max(y_cities) - np.min(y_cities)
    area_coverage = x_range * y_range
    density = len(cities) / area_coverage
    
    print(f"   Number of cities: {len(cities)}")
    print(f"   X range: [{np.min(x_cities):.4f}, {np.max(x_cities):.4f}]")
    print(f"   Y range: [{np.min(y_cities):.4f}, {np.max(y_cities):.4f}]")
    print(f"   Area coverage: {area_coverage:.4f}")
    print(f"   Spatial density: {density:.2f} cities/unit²")
    
    # Calculate distance matrix for complexity analysis
    distance_matrix = np.zeros((len(cities), len(cities)))
    for i in range(len(cities)):
        for j in range(i+1, len(cities)):
            dist = np.sqrt((x_cities[i] - x_cities[j])**2 + (y_cities[i] - y_cities[j])**2)
            distance_matrix[i,j] = distance_matrix[j,i] = dist
    
    # Distance statistics
    distances = distance_matrix[np.triu_indices(len(cities), k=1)]
    mean_distance = np.mean(distances)
    std_distance = np.std(distances)
    min_distance = np.min(distances[distances > 0])
    max_distance = np.max(distances)
    
    print(f"\nDISTANCE MATRIX ANALYSIS:")
    print(f"   Mean inter-city distance: {mean_distance:.4f} ± {std_distance:.4f}")
    print(f"   Min/max distance: [{min_distance:.4f}, {max_distance:.4f}]")
    print(f"   Coefficient of variation: {std_distance/mean_distance:.3f}")
    
    # Calculate theoretical lower bound (MST approximation)
    from scipy.sparse import csr_matrix
    from scipy.sparse.csgraph import minimum_spanning_tree
    
    try:
        mst = minimum_spanning_tree(csr_matrix(distance_matrix)).toarray()
        mst_weight = np.sum(mst)
        print(f"   MST Lower Bound: {mst_weight:.4f}")
    except ImportError:
        print("   scipy not available for MST analysis")

if len(path) > 0:
    # Optimal path analysis
    print(f"\nOPTIMAL PATH ANALYSIS:")
    
    # Calculate total length and segment statistics
    segment_lengths = []
    for i in range(len(path)):
        dx = x_path[(i+1) % len(path)] - x_path[i]
        dy = y_path[(i+1) % len(path)] - y_path[i]
        segment_length = np.sqrt(dx**2 + dy**2)
        segment_lengths.append(segment_length)
    
    path_length = sum(segment_lengths)
    mean_segment = np.mean(segment_lengths)
    std_segment = np.std(segment_lengths)
    
    print(f"   Total path length: {path_length:.4f}")
    print(f"   Mean segment length: {mean_segment:.4f} ± {std_segment:.4f}")
    print(f"   Min/max segment: [{min(segment_lengths):.4f}, {max(segment_lengths):.4f}]")
    
    # Calculate efficiency vs theoretical bounds
    if 'mst_weight' in locals():
        efficiency_vs_mst = mst_weight / path_length
        print(f"   Efficiency vs MST bound: {efficiency_vs_mst:.3f} ({efficiency_vs_mst*100:.1f}%)")
    
    # Crossing detection analysis
    def segments_intersect(p1, p2, p3, p4):
        def orientation(p, q, r):
            val = (q[1] - p[1]) * (r[0] - q[0]) - (q[0] - p[0]) * (r[1] - q[1])
            if val == 0: return 0
            return 1 if val > 0 else 2
        
        o1 = orientation(p1, p2, p3)
        o2 = orientation(p1, p2, p4)
        o3 = orientation(p3, p4, p1)
        o4 = orientation(p3, p4, p2)
        
        return (o1 != o2 and o3 != o4)
    
    crossings = 0
    path_points = list(zip(x_path, y_path))
    n = len(path_points)
    
    for i in range(n):
        for j in range(i+2, n):
            if j == n-1 and i == 0:
                continue
            if segments_intersect(path_points[i], path_points[(i+1) % n],
                                path_points[j], path_points[(j+1) % n]):
                crossings += 1
    
    print(f"   Crossings detected: {crossings}")
    print(f"   Topological quality: {'OPTIMAL' if crossings == 0 else 'SUBOPTIMAL'}")

# Main visualization: TSP solution
if len(cities) > 0 and len(path) > 0:
    fig, ax = plt.subplots(1, 1, figsize=(12, 10))
    
    # Plot cities
    scatter = ax.scatter(x_cities, y_cities, c=range(len(cities)), cmap='viridis', 
                         s=100, alpha=0.8, edgecolors='black', linewidth=1, 
                         label=f'Cities ({len(cities)})', zorder=3)
    
    # Plot optimal path
    ax.plot(x_path, y_path, 'r-', linewidth=3, alpha=0.8, label='Optimal Path', zorder=2)
    ax.plot(x_path, y_path, 'ro', markersize=6, alpha=0.7, zorder=2)
    ax.plot([x_path[-1], x_path[0]], [y_path[-1], y_path[0]], 'r-', linewidth=3, alpha=0.8)
    
    # Highlight start/end point
    ax.plot(x_path[0], y_path[0], 'go', markersize=12, markeredgecolor='black', 
            markeredgewidth=2, label='Start/End', zorder=4)
    
    # Add colorbar
    cbar = plt.colorbar(scatter, ax=ax)
    cbar.set_label('City Index', rotation=270, labelpad=15)
    
    ax.set_xlabel('X Coordinate')
    ax.set_ylabel('Y Coordinate')
    ax.set_title(f'TSP Optimal Path - {len(cities)} Cities\nLength: {path_length:.4f}', 
                 fontweight='bold', fontsize=14)
    ax.legend()
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal')

plt.tight_layout()
plt.savefig('tsp_square_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# Final summary
if len(cities) > 0 and len(path) > 0:
    print(f"\nTSP ANALYSIS SUMMARY:")
    print(f"   Problem solved: {len(cities)} cities")
    print(f"   Path length: {path_length:.4f}")
    print(f"   Crossings: {crossings}")
    if 'mst_weight' in locals():
        print(f"   Gap from lower bound: {((path_length - mst_weight)/mst_weight)*100:.2f}%")

print(f"\nTSP analysis completed\n")
```

```python
# Load and analyze path length evolution data
try:
    evolution_data = np.loadtxt('fitness_evolution.dat')
    generations = evolution_data[:, 0]
    path_lengths = evolution_data[:, 1]  # This is actually L2 path length, not fitness
    
    print(f"Path length evolution data loaded: {len(generations)} generations")
    print(f"Initial path length (L2): {path_lengths[0]:.6f}")
    print(f"Final path length (L2): {path_lengths[-1]:.6f}")
    print(f"Total improvement: {(path_lengths[0] - path_lengths[-1]):.6f}")
    print(f"Relative improvement: {((path_lengths[0] - path_lengths[-1])/path_lengths[0]*100):.2f}%")
    
    data_loaded = True
except FileNotFoundError:
    print("File fitness_evolution.dat not found")
    print("Run the TSP GA algorithm first to generate evolution data")
    data_loaded = False

if data_loaded:
    # Create path length evolution visualization
    fig, ax1 = plt.subplots(1, 1, figsize=(12, 8))
    
    # Main path length evolution plot
    ax1.plot(generations, path_lengths, 'b-', linewidth=2, alpha=0.8, label='Best Path Length (L2)')
    
    # Add moving average for trend analysis
    if len(path_lengths) > 100:
        window_size = max(10, len(path_lengths) // 50)
        moving_avg = np.convolve(path_lengths, np.ones(window_size)/window_size, mode='valid')
        moving_gen = generations[window_size-1:]
        ax1.plot(moving_gen, moving_avg, 'r--', linewidth=2, alpha=0.7,
                label=f'Moving Average (window={window_size})')
    
    # Find and highlight the best (minimum) path length
    best_idx = np.argmin(path_lengths)
    best_generation = generations[best_idx]
    best_path_length = path_lengths[best_idx]
    
    # Add arrow pointing to the best solution
    ax1.annotate(f'Best Solution\nGen: {int(best_generation)}\nLength: {best_path_length:.4f}',
                xy=(best_generation, best_path_length),
                xytext=(best_generation + len(generations)*0.2, best_path_length + (np.max(path_lengths) - np.min(path_lengths))*0.3),
                arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.2', 
                              color='red', lw=2),
                fontsize=12, fontweight='bold',
                bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7),
                ha='center')
    
    # Highlight the best point
    ax1.plot(best_generation, best_path_length, 'ro', markersize=10, 
             markeredgecolor='black', markeredgewidth=2, zorder=5)
    
    ax1.set_xlabel('Generation')
    ax1.set_ylabel('Path Length (L2 Distance)')
    ax1.set_title('TSP Genetic Algorithm - Path Length Evolution', fontweight='bold', fontsize=14)
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('path_length_evolution_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Convergence analysis
    if len(path_lengths) > 50:
        print(f"\nCONVERGENCE ANALYSIS:")
        
        # Find convergence point
        improvement_rate = np.diff(path_lengths) * -1
        convergence_threshold = np.std(improvement_rate) * 0.1
        
        convergence_gen = None
        if len(improvement_rate) > 100:
            for i in range(len(improvement_rate)-50, 50, -1):
                if np.mean(np.abs(improvement_rate[i:i+50])) > convergence_threshold:
                    convergence_gen = i
                    break
        
        if convergence_gen:
            print(f"   Convergence around generation: {convergence_gen}")
            print(f"   Path length at convergence: {path_lengths[convergence_gen]:.6f}")
            print(f"   Further improvement after convergence: {(path_lengths[convergence_gen] - path_lengths[-1]):.6f}")
                
        # Algorithm efficiency metrics
        total_improvement = path_lengths[0] - path_lengths[-1]
        efficient_gens = np.sum(improvement_rate > np.mean(improvement_rate))
        print(f"   Efficient generations (above average): {efficient_gens}/{len(improvement_rate)} ({efficient_gens/len(improvement_rate)*100:.1f}%)")
        
        # Calculate optimization efficiency
        print(f"\nOPTIMIZATION EFFICIENCY:")
        print(f"   Initial L2 distance: {path_lengths[0]:.6f}")
        print(f"   Final L2 distance: {path_lengths[-1]:.6f}")
        print(f"   Total reduction: {total_improvement:.6f}")
        print(f"   Best solution found at generation: {int(best_generation)}")

print(f"\nPath length evolution analysis completed")
```

### Obiettivo
Simulare l'**esperimento di Buffon** per stimare il valore di $\pi$ tramite il lancio di aghi su linee parallele.

<p style="border:2px; border-style:solid; border-color:#E6E6FA; padding: 1em; background-color:#F0F8FF">
<b>Esperimento di Buffon:</b><br>
Un ago di lunghezza $L$ viene lanciato casualmente su un piano rigato con linee parallele distanti $d$ (con $d > L$).
<br><br>
<b>Probabilità di intersezione:</b>
$$P = \frac{2L}{\pi d}$$
<br>
<b>Stima di π:</b> Dopo $N_{thr}$ lanci con $N_{hit}$ intersezioni:
$$\pi = \lim_{N_{thr} \to \infty} \frac{2L N_{thr}}{N_{hit} d}$$
</p>

```python
import numpy as np
import matplotlib.pyplot as plt

# Carico i dati di Buffon (blocchi, stima_pi, errore)
data = np.loadtxt("output1_3.dat")

blocks = data[:,0]     # Numero di blocchi
pi_est = data[:,1]     # Stima di π
error = data[:,2]      # Errore statistico

# Valore teorico
pi_theory = np.pi

# Plot della stima di π
plt.figure(figsize=(10,6))
plt.errorbar(blocks, pi_est, yerr=error, fmt='o-', capsize=3, markersize=4,
             label="Stima di π (Buffon)", color='darkslateblue')
plt.axhline(pi_theory, color='crimson', linestyle='--', linewidth=2,
            label=f"π teorico = {pi_theory:.4f}")
plt.xlabel("Numero di blocchi N", fontsize=12)
plt.ylabel("Stima di π", fontsize=12)
plt.title("Stima di π con l'Esperimento di Buffon", fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Valore finale: π = {pi_est[-1]:.6f} ± {error[-1]:.6f}")
print(f"Valore teorico: π = {pi_theory:.6f}")
print(f"Differenza: {abs(pi_est[-1] - pi_theory):.6f} ({abs(pi_est[-1] - pi_theory)/error[-1]:.1f}σ)")
```

### **Interpretazione dei Risultati**

- La probabilità geometrica che un ago intersechi una linea dipende da π
- La stima migliora con l'aumento del numero di lanci

### Obiettivo
Valutare opzioni Call e Put europee utilizzando simulazioni Monte Carlo e confrontarle con i prezzi teorici del modello **Black-Scholes**.

<p style="border:2px; border-style:solid; border-color:#E6E6FA; padding: 1em; background-color:#F0F8FF">
<b>Introduzione Teorica:</b><br>
Il valore intrinseco delle opzioni alla scadenza è dato da:
$$\text{Call: } \max(S_T - K, 0) \quad \text{Put: } \max(K - S_T, 0)$$
dove $S_T$ indica il prezzo dell'asset al tempo $T$. Per riportarlo al tempo attuale, il profitto va scontato: infatti, va moltiplicato per un esponenziale con esponente negativo $e^{-rT}$, che tiene conto del fatto che il valore del denaro non è costante nel tempo.
<br><br>
Si è assunto che l'evoluzione temporale del prezzo segua un moto browniano geometrico (GBM):
$$dS_t = rS_t dt + \sigma S_t dW_t$$
Dove:
<br>• $r$ rappresenta il tasso di interesse privo di rischio
<br>• $\sigma$ esprime la volatilità del titolo  
<br>• $dW_t$ è un processo di Wiener standard con incrementi $dW_t \sim \mathcal{N}(0, dt)$
<br><br>
Si è valutato il prezzo dell'opzione tramite simulazione Monte Carlo, con blocchi, e per un totale di repliche del processo stocastico.
<br><br>
<b>1. Campionamento Diretto del Prezzo Finale</b><br>
Si è generato direttamente il valore $S_T$ mediante la formula chiusa:
$$S_T = S_0 \exp\left[\left(r - \frac{\sigma^2}{2}\right)T + \sigma\sqrt{T}\, Z\right]$$
con $Z \sim \mathcal{N}(0,1)$.
<br><br>
<b>2. Campionamento Discretizzato del Percorso</b><br>
Si è suddiviso l'intervallo $[0,T]$ in $N$ sottointervalli, simulando l'evoluzione incrementale:
$$S_{t+\Delta t} = S_t \exp\left[\left(r - \frac{\sigma^2}{2}\right)\Delta t + \sigma\sqrt{\Delta t}\, Z\right]$$
Questa approssimazione permette di catturare la dinamica temporale del prezzo, risultando particolarmente utile per opzioni path-dependent.
<br><br>
<b>Pricing delle opzioni:</b><br>
• <b>Call</b>: $C = e^{-rT} \mathbb{E}[\max(S_T - K, 0)]$<br>
• <b>Put</b>: $P = e^{-rT} \mathbb{E}[\max(K - S_T, 0)]$<br>
<br>
dove $K$ è lo strike price, $T$ la scadenza, $S_T$ il prezzo finale dell'asset.
</p>

```python
import numpy as np
import matplotlib.pyplot as plt

### **Analisi del Pricing Monte Carlo vs Black-Scholes**

# Carica dati dal file
data = np.loadtxt("option_prices.dat", skiprows=1)

# Estrai colonne
block = data[:, 0]
call_price = data[:, 1]
call_error = data[:, 2]
put_price = data[:, 3]
put_error = data[:, 4]

# Valori teorici Black-Scholes
bs_call = 14.975790778311286
bs_put = 5.4595325819072364

print("=== PRICING DI OPZIONI: MONTE CARLO vs BLACK-SCHOLES ===")
print(f"Numero di blocchi analizzati: {len(block)}")
print()

# Risultati finali
print("PREZZI FINALI:")
print(f"Call Option:")
print(f"  Monte Carlo:    {call_price[-1]:.6f} ± {call_error[-1]:.6f}")
print(f"  Black-Scholes:  {bs_call:.6f}")
print(f"  Differenza:     {abs(call_price[-1] - bs_call):.6f}")
print()
print(f"Put Option:")
print(f"  Monte Carlo:    {put_price[-1]:.6f} ± {put_error[-1]:.6f}")
print(f"  Black-Scholes:  {bs_put:.6f}")  
print(f"  Differenza:     {abs(put_price[-1] - bs_put):.6f}")
print()

# Verifica Put-Call Parity
S0, K, r, T = 100, 100, 0.1, 1.0
parity_theoretical = S0 - K * np.exp(-r * T)
parity_mc = call_price[-1] - put_price[-1]
print(f"PUT-CALL PARITY CHECK:")
print(f"  C - P (MC):      {parity_mc:.6f}")
print(f"  S₀ - K·e^(-rT):  {parity_theoretical:.6f}")
print(f"  Differenza:      {abs(parity_mc - parity_theoretical):.6f}")

# Accuratezza relativa
call_accuracy = abs(call_price[-1] - bs_call) / bs_call * 100
put_accuracy = abs(put_price[-1] - bs_put) / bs_put * 100
print(f"\nACCURATEZZA RELATIVA:")
print(f"Call Option: {call_accuracy:.4f}%")
print(f"Put Option:  {put_accuracy:.4f}%")

# Plot CALL Option
plt.figure(figsize=(10,6))
plt.errorbar(block, call_price, yerr=call_error, fmt='o', capsize=2, label='Call MC')
plt.axhline(bs_call, color='red', linestyle='--', label='Call Black-Scholes')
plt.xlabel('Block')
plt.ylabel('Call Price')
plt.title('Call Option Price: Monte Carlo (One step) vs Black-Scholes')
plt.grid(True)
plt.legend()
plt.show()

# Plot PUT Option
plt.figure(figsize=(10,6))
plt.errorbar(block, put_price, yerr=put_error, fmt='o', capsize=2, color='orange', label='Put MC')
plt.axhline(bs_put, color='red', linestyle='--', label='Put Black-Scholes')
plt.xlabel('Block')
plt.ylabel('Put Price')
plt.title('Put Option Price: Monte Carlo (One step) vs Black-Scholes')
plt.grid(True)
plt.legend()
plt.show()
```

### **Interpretazione dei Risultati**

L'analisi del pricing di opzioni rivela aspetti fondamentali del metodo Monte Carlo applicato alla finanza quantitativa:

**Put-Call Parity:**
La relazione $C - P = S_0 - Ke^{-rT}$ è una proprietà fondamentale che deve essere rispettata:
- Verifica l'arbitraggio: se violata, esisterebbero opportunità di profitto senza rischio
- Conferma la coerenza tra i prezzi delle due opzioni

**Convergenza Monte Carlo:**
- Gli errori diminuiscono come $1/\sqrt{N}$ con il numero di simulazioni
- Convergenza più lenta rispetto ai metodi deterministici, ma indipendente dalla dimensionalità

**Vantaggi del Metodo One-Step:**
- **Efficienza computazionale**: una sola generazione di numeri casuali per simulazione
- **Semplicità di implementazione**: direttamente dal modello Black-Scholes
- **Accuratezza**: ottimo accordo con la teoria per opzioni europee vanilla

### Obiettivo
Simulare l'**evoluzione temporale** di un sistema Lennard-Jones da una configurazione altamente ordinata (cristallo FCC: bassa entropia) verso l'equilibrio termodinamico, dimostrando la convergenza della distribuzione delle velocità verso la distribuzione Maxwell-Boltzmann.

### Background Fisico

Il sistema parte da una **configurazione a bassa entropia**:
- **Posizioni iniziali**: cristallo FCC (Face-Centered Cubic) che occupa metà del box
- **Velocità iniziali**: distribuzione δ-funzione p_δ(v*) = δ(v* - v_T*) con velocità assegnata a una sola componente

La **termalizzazione spontanea** avviene attraverso:
1. **Redistribuzione energetica**: l'energia si ridistribuisce tra componenti cinetiche e potenziali
2. **Scambi energetici**: le interazioni Lennard-Jones causano scambi tra particelle
3. **Crescita dell'entropia**: il sistema evolve verso stati di maggiore probabilità
4. **Convergenza Maxwell-Boltzmann**: la distribuzione δ si "allarga" verso la forma teorica

<p style="border:2px; border-style:solid; border-color:#E6E6FA; padding: 1em; background-color:#F0F8FF">
<b>Parametri di simulazione:</b><br>
• Modello: gas Lennard-Jones rarefatto (ρ* ≈ 0.05, T* ≈ 2.0)<br>
• Cut-off radius: r_c = 5.0<br>
<b>File di analisi:</b> pofv.dat, temperature.dat, configurazioni spaziali
</p>


```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
from math import pi

# ---------------------- UTILITY: path finder ----------------------
def find_output_file(fname):
    candidates = [os.path.join("OUTPUT", fname), os.path.join("..", "OUTPUT", fname),
                  os.path.join(".", fname), os.path.join("..", fname)]
    for c in candidates:
        if os.path.isfile(c):
            return c
    raise FileNotFoundError(f"Non ho trovato {fname} in {candidates}. Assicurati che il simulatore abbia scritto il file.")

pofv_path = find_output_file("pofv.dat")
temp_path = find_output_file("temperature.dat")

print("=== CONVERGENZA ALLA DISTRIBUZIONE DI MAXWELL-BOLTZMANN ===")
print("File utilizzati:", pofv_path, " e ", temp_path)

# ---------------------- CARICA pofv.dat ----------------------
data = np.loadtxt(pofv_path)
if data.ndim == 1:
    data = data.reshape((1, -1))

v_centers_unique = np.unique(data[:,0])
nbins = len(v_centers_unique)
nlines = data.shape[0]
nblocks = nlines // nbins

print(f"Righe totali: {nlines}, bin unici: {nbins}, blocchi: {nblocks}")

# ricostruisci array blocks x bins x 4 (v, average, sum_average, err)
blocks_data = data.reshape((nblocks, nbins, data.shape[1]))
v_centers = blocks_data[0,:,0]

# ---------------------- CARICA temperature.dat ----------------------
temp_data = np.loadtxt(temp_path)
if temp_data.ndim == 1:
    temp_data = temp_data.reshape((1, -1))

if temp_data.shape[1] >= 2:
    T_block = temp_data[:,1]  # temperatura effettiva per blocco
else:
    T_block = temp_data[:,0]

if len(T_block) != nblocks:
    nblocks_common = min(len(T_block), nblocks)
    T_block = T_block[:nblocks_common]
    blocks_data = blocks_data[:nblocks_common,:,:]
    nblocks = nblocks_common

# ---------------------- Funzione Maxwell-Boltzmann teorica ----------------------
def maxwell_boltzmann_3d(v, T):
    """Distribuzione di Maxwell-Boltzmann 3D in unità ridotte LJ"""
    pref = (1.0 / ((2.0 * np.pi * T)**1.5))
    return pref * 4.0 * np.pi * v**2 * np.exp(-v**2 / (2.0 * T))

def normalize_pdf(v, pdf):
    """Normalizza numericamente una PDF"""
    area = np.trapz(pdf, v)
    if area == 0:
        return pdf, area
    return pdf/area, area


# ---------------------- EVOLUZIONE BLOCCHI SPECIFICI ----------------------
# Blocchi da plottare: 1, 5, 10, 15, 20
target_blocks = [1, 5, 10, 15, 20]
# Converti in indici (1-based -> 0-based)
block_indices = [b-1 for b in target_blocks if b-1 < nblocks]
actual_blocks = [i+1 for i in block_indices]  # blocchi effettivamente disponibili

print(f"Blocchi da analizzare: {actual_blocks}")

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Evoluzione della Distribuzione delle Velocità: da δ-funzione a Maxwell-Boltzmann', 
             fontsize=16, fontweight='bold')
colors = ['red', 'orange', 'green', 'blue', 'purple']
v_theo = np.linspace(0, v_centers.max()*1.1, 300)

# Plot dei singoli blocchi
for idx, (block_idx, color, block_num) in enumerate(zip(block_indices, colors, actual_blocks)):
    if idx < 5:  # massimo 5 blocchi
        row = idx // 3
        col = idx % 3
        ax = axes[row, col]
        
        # Dati del blocco
        block = blocks_data[block_idx,:,:]
        v = block[:,0]
        pdf_raw = block[:,1]
        err_raw = block[:,3]
        pdf, area = normalize_pdf(v, pdf_raw)
        
        if area != 0:
            err = err_raw / area
        else:
            err = err_raw * 0.0
        
        T_eff = T_block[block_idx]
        
        # Plot istogramma
        ax.bar(v, pdf, width=(v[1]-v[0])*0.8, alpha=0.7, color=color, 
               edgecolor='black', linewidth=0.5, label=f'Simulazione MD')
        
        # Plot barre di errore
        ax.errorbar(v, pdf, yerr=err, fmt='none', ecolor='black', capsize=2, alpha=0.8)
        
        # Maxwell-Boltzmann teorica
        p_theory = maxwell_boltzmann_3d(v_theo, T_eff)
        ax.plot(v_theo, p_theory, 'k--', linewidth=2, label='Maxwell-Boltzmann teorica')
        
        ax.set_xlabel('Velocità v*')
        ax.set_ylabel('p(v*)')
        ax.set_title(f'BLOCCO {block_num}\nT* = {T_eff:.3f}', fontweight='bold', fontsize=12)
        ax.legend(fontsize=10)
        ax.grid(True, alpha=0.3)
        ax.set_xlim(0, 4)
        
        if block_num == 1:
            ax.text(0.6, 0.9, 'δ-like\n(bassa entropia)', transform=ax.transAxes, 
                   bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7),
                   fontsize=10, ha='center')
        elif block_num == actual_blocks[-1]:
            ax.text(0.6, 0.9, 'Maxwell-Boltzmann\n(alta entropia)', transform=ax.transAxes,
                   bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen", alpha=0.7),
                   fontsize=10, ha='center')

# Rimuovi il subplot vuoto (2,2)
fig.delaxes(axes[1, 2])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

def find_file(name):
    candidates = [os.path.join("OUTPUT", name), os.path.join("..","OUTPUT", name),
                  name, os.path.join("..", name)]
    for c in candidates:
        if os.path.isfile(c):
            return c
    raise FileNotFoundError(f"File {name} non trovato nei percorsi: {candidates}")

def load_progressive_file(path):
    data = np.loadtxt(path)
    if data.ndim == 1:
        data = data.reshape((1,-1))
    blocks = data[:,0].astype(int)
    avg_prog = data[:,2]
    err_prog = data[:,3]
    return blocks, avg_prog, err_prog

# ---- trova i file ----
try:
    kin_f = find_file("kinetic_energy.dat")
    pot_f = find_file("potential_energy.dat")
    temp_f = find_file("temperature.dat")
except FileNotFoundError as e:
    print(e)
    raise SystemExit(1)

print("Uso file:", kin_f, pot_f, temp_f)

# ---- carica dati ----
blk_k, kin_prog, kin_err = load_progressive_file(kin_f)
blk_p, pot_prog, pot_err = load_progressive_file(pot_f)
blk_t, temp_prog, temp_err = load_progressive_file(temp_f)

# allinea numero di blocchi al minimo comune (sicurezza)
nblocks = min(len(blk_k), len(blk_p), len(blk_t))
blk = blk_k[:nblocks]
kin_prog = kin_prog[:nblocks]; kin_err = kin_err[:nblocks]
pot_prog = pot_prog[:nblocks]; pot_err = pot_err[:nblocks]
temp_prog = temp_prog[:nblocks]; temp_err = temp_err[:nblocks]

# ---- total energy e propagazione errore (assumiamo indipendenza) ----
total_prog = kin_prog + pot_prog
total_err = np.sqrt( kin_err**2 + pot_err**2 )

# ---- plot Energies (K, P, Total) ----
plt.figure(figsize=(9,6))
plt.errorbar(blk, kin_prog, yerr=kin_err, fmt='o-', label='Kinetic Energy per particle', capsize=3)
plt.errorbar(blk, pot_prog, yerr=pot_err, fmt='s-', label='Potential Energy per particle', capsize=3)
plt.errorbar(blk, total_prog, yerr=total_err, fmt='^-', label='Total Energy per particle', capsize=3, color='black')
plt.xlabel('Block index')
plt.ylabel('Energy per particle (reduced units)')
plt.title('Kinetic, Potential and Total Energy (progressive averages)')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()
plt.savefig('energy_progressive.png', dpi=200)
print("Salvato: energy_progressive.png")
plt.show()

# ---- plot Temperature ----
plt.figure(figsize=(9,5))
plt.errorbar(blk, temp_prog, yerr=temp_err, fmt='o-', label='Temperature (progressive)', capsize=3)
# prova a leggere TEMP da ../INPUT/input.dat e disegnare linea orizzontale se esiste
input_paths = [os.path.join("INPUT","input.dat"), os.path.join("..","INPUT","input.dat"), "input.dat", os.path.join("..","input.dat")]
temp_input_val = None
for ip in input_paths:
    if os.path.isfile(ip):
        try:
            with open(ip,'r') as f:
                for line in f:
                    toks = line.split()
                    if len(toks) >= 2 and toks[0].upper() == "TEMP":
                        try:
                            temp_input_val = float(toks[1])
                            print("Trovato TEMP in", ip, "=", temp_input_val)
                        except:
                            pass
                        break
        except:
            pass
        if temp_input_val is not None:
            break

if temp_input_val is not None:
    plt.axhline(temp_input_val, color='red', linestyle='--', label=f"T input = {temp_input_val:.3f}")

plt.xlabel('Block index')
plt.ylabel('Temperature (reduced units)')
plt.title('Temperature evolution (progressive averages)')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()
plt.savefig('temperature_progressive.png', dpi=200)
print("Salvato: temperature_progressive.png")
plt.show()
```

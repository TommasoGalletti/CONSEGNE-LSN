### Obiettivo
Calcolare l'integrale tramite metodi Monte Carlo:
$$I = \int_0^1 \frac{\pi}{2}\cos\left(\frac{\pi x}{2}\right) dx = 1$$

Confrontando due approcci:
1. **Campionamento uniforme** in $[0,1]$
2. **Importance sampling** con distribuzione non uniforme

<p style="border:2px; border-style:solid; border-color:#E6E6FA; padding: 1em; background-color:#F0F8FF">
<b>Metodo Monte Carlo per integrazione:</b><br>
Per calcolare $I = \int_a^b f(x) dx$ con $N$ campioni:
<br><br>
<b>Campionamento uniforme:</b> $I \approx \frac{b-a}{N} \sum_{i=1}^N f(x_i)$ con $x_i \sim U[a,b]$
<br><br>
<b>Importance sampling:</b> $I \approx \frac{1}{N} \sum_{i=1}^N \frac{f(x_i)}{p(x_i)}$ con $x_i \sim p(x)$
<br>
dove $p(x)$ è una distribuzione che approssima $|f(x)|$.
</p>

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("output.dat", skiprows=1)

N = data[:, 0]  # Numero di blocchi inclusi progressivamente
mean_uniform = data[:, 1]
error_uniform = data[:, 2]
mean_importance = data[:, 3]
error_importance = data[:, 4]

# Valore atteso dell'integrale (da calcolo analitico)
I_exact = 1.0

print("=== INTEGRAZIONE MONTE CARLO ===")
print(f"Integrale da calcolare: ∫₀¹ (π/2)cos(πx/2) dx = {I_exact}")
print(f"Numero di blocchi analizzati: {len(N)}")
print()

# Risultati finali
print("RISULTATI FINALI:")
print(f"Campionamento uniforme:   I = {mean_uniform[-1]:.6f} ± {error_uniform[-1]:.6f}")
print(f"Importance sampling:      I = {mean_importance[-1]:.6f} ± {error_importance[-1]:.6f}")
print(f"Valore teorico:           I = {I_exact:.6f}")
print()
print(f"Errore relativo uniforme:    {abs(mean_uniform[-1]-I_exact)/I_exact*100:.4f}%")
print(f"Errore relativo importance:  {abs(mean_importance[-1]-I_exact)/I_exact*100:.4f}%")

# Grafico uniforme
plt.figure(figsize=(10,5))
plt.errorbar(N, mean_uniform, yerr=error_uniform, fmt="o-", label="Uniform Sampling")
plt.axhline(I_exact, color="r", linestyle="dashed", label="Valore esatto")
plt.xlabel("Numero di blocchi")
plt.ylabel("Stima dell'integrale")
plt.title("Convergenza del metodo Monte Carlo - Uniform Sampling")
plt.legend()
plt.grid()
plt.show()

# Grafico importance sampling
plt.figure(figsize=(10,5))
plt.errorbar(N, mean_importance, yerr=error_importance, fmt="o-", label="Importance Sampling")
plt.axhline(I_exact, color="r", linestyle="dashed", label="Valore esatto")
plt.xlabel("Numero di blocchi")
plt.ylabel("Stima dell'integrale")
plt.title("Convergenza del metodo Monte Carlo - Importance Sampling")
plt.legend()
plt.grid()
plt.show()
```

```python
# uniform e importance sampling
plt.figure(figsize=(12,6))
plt.errorbar(N, mean_uniform, yerr=error_uniform, fmt="o-", label="Uniform Sampling", alpha=0.8)
plt.errorbar(N, mean_importance, yerr=error_importance, fmt="s-", label="Importance Sampling", alpha=0.8)
plt.axhline(I_exact, color="r", linestyle="dashed", label="Valore esatto")
plt.xlabel("Numero di blocchi")
plt.ylabel("Stima dell'integrale")
plt.title("Confronto Metodi Monte Carlo")
plt.legend()
plt.grid()
plt.show()
```

### **Interpretazione dei Risultati**

**Campionamento Uniforme:**
- Metodo standard che campiona uniformemente in $[0,1]$
- Convergenza stabile ma potenzialmente più lenta
- Semplicità di implementazione

**Importance Sampling:**  
- Utilizza una distribuzione $p(x)$ che privilegia regioni dove $f(x)$ è grande
- Può ridurre significativamente la varianza se $p(x) \propto |f(x)|$
- Richiede conoscenza a priori della funzione integranda

**Considerazioni Teoriche:**
La convergenza Monte Carlo segue la legge $\sigma \propto 1/\sqrt{N}$, indipendentemente dalla dimensionalità del problema, che è il principale vantaggio rispetto ai metodi deterministici.
